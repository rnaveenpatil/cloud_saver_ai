```python
import pandas as pd

df = pd.read_csv("final_removed.csv")

df_sorted = (
    df
    .sort_values(
        by=["collection_id", "instance_index", "time"],
        ascending=[True, True, True]   # 👈 time ASCENDING
    )
    .reset_index(drop=True)
)

print("Sorted shape:", df_sorted.shape)
```

```python
df_sorted.groupby(
    ["collection_id", "instance_index"]
)["time"].is_monotonic_increasing.value_counts()
```

```python
df_sorted.to_csv("final_colums_sorted.csv", index=False)
```
