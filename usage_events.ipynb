```python
import pandas as pd

events = pd.read_csv(
    "/home/mudda/cluster2011/task_events/merged_task_events.csv",
    header=None,
    low_memory=False
)

events.columns = [
    "timestamp","missing_info","job_id","task_index","machine_id",
    "event_type","user","scheduling_class","priority",
    "cpu_request","ram_request","disk_request",
    "different_machine_constraint"
]

failed_tasks = events.loc[
    events["event_type"] == 3,
    ["job_id", "task_index"]
].drop_duplicates()

failed_tasks["failed"] = 1

print("Failed tasks ready ✅")
```

```python
import pandas as pd

# Output file
output_file = "/home/mudda/cluster2011/cluster2011_final_labeled.csv"

chunk_size = 200000   # 🔥 safer for low RAM
first_chunk = True

for chunk in pd.read_csv(
        "/home/mudda/cluster2011/task_usage/merged_task_usage.csv",
        header=None,
        chunksize=chunk_size,
        engine="python",
        on_bad_lines="skip"
    ):

    # ✅ Assign all 20 columns
    chunk.columns = [
        "start_time",                 # 0
        "end_time",                   # 1
        "job_id",                     # 2
        "task_index",                 # 3
        "machine_id",                 # 4
        "cpu_rate",                   # 5
        "canonical_memory_usage",     # 6
        "assigned_memory_usage",      # 7
        "unmapped_page_cache",        # 8
        "total_page_cache",           # 9
        "maximum_memory_usage",       # 10
        "disk_io_time",               # 11
        "local_disk_space_usage",     # 12
        "maximum_cpu_rate",           # 13
        "maximum_disk_io_time",       # 14
        "cpi",                        # 15
        "mai",                        # 16
        "sample_portion",             # 17
        "aggregation_type",           # 18
        "extra_field"                 # 19
    ]

    # 🔥 SAFE numeric conversion (prevents crash)
    chunk["job_id"] = pd.to_numeric(chunk["job_id"], errors="coerce")
    chunk["task_index"] = pd.to_numeric(chunk["task_index"], errors="coerce")

    # Drop corrupted rows
    chunk = chunk.dropna(subset=["job_id", "task_index"])

    # Convert properly
    chunk["job_id"] = chunk["job_id"].astype("int64")
    chunk["task_index"] = chunk["task_index"].astype("int64")

    # Merge with failed_tasks
    chunk = chunk.merge(
        failed_tasks,
        on=["job_id", "task_index"],
        how="left"
    )

    chunk["failed"] = chunk["failed"].fillna(0).astype("int8")

    # Write safely
    chunk.to_csv(
        output_file,
        mode="w" if first_chunk else "a",
        header=first_chunk,
        index=False
    )

    first_chunk = False

    # 🔥 Free RAM manually
    del chunk

print("✅ DONE — Final labeled dataset created safely.")
```

```python
import pandas as pd

input_file = "/home/mudda/cluster2011/cluster2011_final_labeled.csv"
output_file = "/home/mudda/cluster2011/final.csv"

chunk_size = 200000
first_chunk = True

for chunk in pd.read_csv(
        input_file,
        chunksize=chunk_size,
        engine="python",
        on_bad_lines="skip"
    ):

    # Rename columns if needed (adjust if names differ in your file)
    chunk = chunk.rename(columns={
        "start_time": "timestamp",
        "cpu_rate": "cpu_usage",
        "canonical_memory_usage": "memory_usage",
        "assigned_memory_usage": "assigned_memory",
        "cpu_request": "requested_cpu",
        "ram_request": "requested_memory",
        "cpi": "cycles_per_instruction",
        "mai": "memory_accesses_per_instruction",
        "total_page_cache": "page_cache_memory"
    })

    # Keep only required columns (ignore if missing)
    required_columns = [
        "timestamp",
        "cpu_usage",
        "memory_usage",
        "assigned_memory",
        "requested_cpu",
        "requested_memory",
        "cycles_per_instruction",
        "memory_accesses_per_instruction",
        "page_cache_memory",
        "disk_io_time",
        "priority",
        "scheduling_class",
        "machine_cpu_capacity",
        "machine_memory_capacity",
        "failed"
    ]

    # Keep only columns that exist
    chunk = chunk[[col for col in required_columns if col in chunk.columns]]

    # 🔥 Remove rows with missing values
    chunk = chunk.dropna()

    # Save safely
    chunk.to_csv(
        output_file,
        mode="w" if first_chunk else "a",
        header=first_chunk,
        index=False
    )

    first_chunk = False
    del chunk

print("✅ Cleaned dataset saved successfully.")
```

```python
import pandas as pd

input_file = "/home/mudda/cluster2011/cluster2011_final_labeled.csv"
output_file = "/home/mudda/cluster2011/final_2.csv"

chunk_size = 200000
first_chunk = True

# Columns you want to keep
required_columns = [
    "timestamp",
    "job_id",
    "task_index",
    "machine_id",
    "cpu_usage",
    "memory_usage",
    "assigned_memory",
    "requested_cpu",
    "requested_memory",
    "page_cache_memory",
    "disk_io_time",
    "local_disk_usage",
    "cycles_per_instruction",
    "memory_accesses_per_instruction",
    "priority",
    "scheduling_class",
    "machine_cpu_capacity",
    "machine_memory_capacity",
    "machine_event_type",
    "failed"
]

for chunk in pd.read_csv(
        input_file,
        chunksize=chunk_size,
        engine="python",
        on_bad_lines="skip"
    ):

    # Keep only required columns that exist
    chunk = chunk[[col for col in required_columns if col in chunk.columns]]

    # 🔥 Convert numeric columns safely
    numeric_cols = [
        "timestamp","job_id","task_index","machine_id",
        "cpu_usage","memory_usage","assigned_memory",
        "requested_cpu","requested_memory",
        "page_cache_memory","disk_io_time",
        "local_disk_usage","cycles_per_instruction",
        "memory_accesses_per_instruction",
        "priority","scheduling_class",
        "machine_cpu_capacity","machine_memory_capacity",
        "machine_event_type","failed"
    ]

    for col in numeric_cols:
        if col in chunk.columns:
            chunk[col] = pd.to_numeric(chunk[col], errors="coerce")

    # Remove corrupted rows
    chunk = chunk.dropna()

    # Save safely
    chunk.to_csv(
        output_file,
        mode="w" if first_chunk else "a",
        header=first_chunk,
        index=False
    )

    first_chunk = False
    del chunk

print("✅ Cleaned dataset saved successfully.")
```
